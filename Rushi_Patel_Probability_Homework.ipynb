The CLT says that in many situations when independent variables are summed up their sum tends towards a normal distribution. In this example as the amount of N's increase the mean of m computations approach a bell curve.

```python
# Type the Python code here
import numpy as np
import matplotlib.pyplot as plt

m = 10000

for N in [1, 5, 10, 20]:
  x = np.random.normal(0,1, m)
  mean = np.mean(x)
  plt.hist(x, 1000)
  plt.show()
```

```python
# Type the Python code here

# Part 1
trials = [0, 5, 15, 50, 500]
out = []
pH = 0.6
for n in trials:
  out.append(np.random.binomial(n,pH))
  print(out)

# Part 2
data = []
for i in range(0, 5):
  if(i == 0):
    data.append(0)
    continue
  data.append(out[i]/trials[i])
print(data)
print(np.mean(data))

plt.scatter(trials,data)
```
